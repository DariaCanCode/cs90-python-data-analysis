# Data Analysis and Visualization: EZ mode

## Get the Data (from a server)

```python
import requests

r = requests.get("https://pokeapi.co/api/v2/pokemon/151/")

data = r.json()

data['stats']
```

## Store the Data (JSON File)

```python
# get the data for the first 151 pokemon
all_data = []
for i in range(1,152):
    r = requests.get(f"https://pokeapi.co/api/v2/pokemon/{i}/")
    data = r.json()
    all_data.append(data)

import json

with open("./pokemon.json", "w") as f:
    json.dump(all_data, f, indent=2)
```

## Get the Data (JSON File)

```python
all_pokemon = None
with open("./pokemon.json") as f:
    all_pokemon = json.load(f)

all_pokemon[0]['name']
```

## Transform the Data

```python
import pandas as pd

columns = ["name", "attack", "defense", "sp_attack", "sp_defense", "speed", "hp"]

pokemon_table = []
for pokemon in all_pokemon:
    row = {
        'name': pokemon['name'],
        'attack': pokemon['stats'][4]['base_stat'],
        'defense': pokemon['stats'][3]['base_stat'],
        'sp_attack': pokemon['stats'][2]['base_stat'],
        'sp_defense': pokemon['stats'][1]['base_stat'],
        'speed': pokemon['stats'][0]['base_stat'],
        'hp': pokemon['stats'][5]['base_stat']
    }
    pokemon_table.append(row)

pokemon_df = pd.DataFrame(pokemon_table, columns=columns)

pokemon_df
```

## Store the Data (CSV File)

```python
pokemon_df.to_csv("./pokemon_stats.csv")
```

## Read the Data (From CSV)

```python
pokemon_df = pd.read_csv("./pokemon_stats.csv", index_col=0)
pokemon_df
```

## Analyze the Data

```python
# Sort
pokemon_sorted = pokemon_df.sort_values(by="attack", ascending=False)
pokemon_sorted
```

```python
# Slice
top_10 = pokemon_sorted[['name', 'attack']][:10]
top_10
```

```python
labels = top_10['name'].values
attack = top_10['attack'].values
print(labels)
print(attack)
```

## Visualize the Data

```python
import matplotlib.pyplot as plt
```

```python
plt.barh(y=[i for i in range(len(labels))], width=attack, tick_label=labels)
plt.show()
```
